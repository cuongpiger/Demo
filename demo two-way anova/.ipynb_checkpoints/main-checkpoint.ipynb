```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df = pd.DataFrame({
    'water': np.repeat(('daily', 'weekly'), 15),
    'sun': np.tile(np.repeat(('low', 'med', 'high'), 5), 2),
    'height': (6, 6, 6, 5, 6, 5, 5, 6, 4, 5, 
               6, 6, 7, 8, 7, 3, 4, 4, 4, 5,
               4, 4, 4, 4, 4, 5, 6, 6, 7, 8)
})

df
```

```python
model = ols('height ~ C(water) + C(sun) + C(water):C(sun)', data=df).fit()
sm.stats.anova_lm(model, typ=2)
```

* Vì **p-value** của ***sun** và ***water*** đều nhỏ hơn $0.05$, điều này có nghĩa là cả hai yếu tố này đều có ý nghĩa về mặt thống kê ảnh hưởng đến ***height*** của cây.
* Và **p-value** của ***water:sun*** là $0.120667$, lớn hơn $0.05$, điều này có nghĩa là ***water*** và ***sun*** không có tương tác qua lại nào lẫn nhau.
